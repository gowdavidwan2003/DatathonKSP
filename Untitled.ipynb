```python
df6 = df[pd.to_numeric(df['Latitude'], errors='coerce').notnull()]
df6 = df6[pd.to_numeric(df['Longitude'], errors='coerce').notnull()]
```

```python
# Step 1: Data Preparation
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

# Assuming 'df' is your dataframe with 'latitude' and 'longitude' columns
X = df6[['Latitude', 'Longitude']]

# Step 2: Distance Calculation
# If X is in degrees, you need to convert it to radians
#X = np.radians(X)

# Step 3: Clustering with DBSCAN
# Scale the data

X = StandardScaler().fit_transform(X)
# Set up and fit the DBSCAN model
dbscan = DBSCAN(eps=0.5, min_samples=10, metric='haversine').fit(X)

# Step 4: Visualization
import folium
map_clusters = folium.Map(location=[13, 77.6], zoom_start=11)
```

```python
labels = dbscan.labels_
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
sscore = silhouette_score(X, labels)
print("Silhouette Score:", sscore)
```
