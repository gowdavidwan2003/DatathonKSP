```python
import re
def clean_distance(distance):
    try:
        distance = re.sub(r'[^0-9]', ' ', distance)
        return distance.strip()
    except:
        return 0
```

```python
df['Cleaned_Distance'] = df['Distance from PS'].apply(clean_distance)
```

```python
plt.figure(figsize=(12, 6))
df['Cleaned_Distance'].value_counts().nlargest(10).plot(kind='bar', color='skyblue')
plt.title('Distribution of Distances from police station ')
plt.xlabel('Section')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```
